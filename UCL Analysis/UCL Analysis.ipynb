```python
possession_data = {
    'Team' : ['Barcelona', 'Man City', 'Sevilla','Bayern','Real Madrid'],
    'Possession' : [59.1, 58.2, 57.3,57.2,56.2],
}

possession_df = pd.DataFrame(possession_data) 



ratings_data={'Team':['Man City','Bayern','Juventus','Chelsea','Madrid'],
'Ratings' : [7.01, 6.99, 6.91,6.9,6.84]}

ratings_df = pd.DataFrame(ratings_data) 



aerial_duels_data={'Team':['Man City','Juventus','Atalanta','Lokomotiv Moscow','Sevilla'],
'Aerial Duels' : [65.6,63.2, 60.7,59.8,56.6]}

aerial_duels_data_df = pd.DataFrame(aerial_duels_data) 




shots_per_game = {'Team':['Barcelona','Salzburg','Bayern','Sevilla','Juventus'],
'Shots per Game' : [18, 17.2, 16.5,16.1,15.5]}

shots_per_game_df = pd.DataFrame(shots_per_game) 
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 10))

fig.suptitle('UCL 20/21 STATS',fontweight ="bold")

# sns.boxplot(ax=axes[0, 0], data=possession_df, x='team', y='possession',palette='Reds_r')
sns.barplot(ax=axes[0, 0],x=possession_df['Team'], y=possession_df['Possession'], palette='Reds_r')
axes[0,0].set_title('Possession',fontweight ="bold")
axes[0,0].set(xlabel=None,ylabel=None)
axes[0,0].axes.yaxis.set_visible(False)
sns.barplot(ax=axes[0, 1],x=ratings_df['Team'], y=ratings_df['Ratings'], palette='Blues_r')
axes[0,1].set_title('Ratings',fontweight ="bold")
axes[0,1].set(xlabel=None,ylabel=None)
axes[0,1].axes.yaxis.set_visible(False)
sns.barplot(ax=axes[1, 1],x=shots_per_game_df['Team'], y=shots_per_game_df['Shots per Game'], palette='Greens_r')
axes[1,1].set_title('Shots per Game',fontweight ="bold")
axes[1,1].set(xlabel=None,ylabel=None)
axes[1,1].axes.yaxis.set_visible(False)
sns.barplot(ax=axes[1, 0],x=aerial_duels_data_df['Team'], y=aerial_duels_data_df['Aerial Duels'], palette='Greys_r')
axes[1,0].set_title('Aerial Duels',fontweight ="bold")
axes[1,0].set(xlabel=None,ylabel=None)
axes[1,0].axes.yaxis.set_visible(False)
fig.savefig("output.png")
```

```python
# import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# load dataset
tips = sns.load_dataset("tips")

# set the figure size
plt.figure(figsize=(14, 14))
data = {
    'Player' : ['E. Haaland', 'K. Mbappé', 'O. Giroud','M. Salah','M. Rashford','K. Benzema','Y. En-Nesyri','Á. Morata','Neymar','R. Lewandowski'],
    'Player Goals/Team Goals' : [10, 8, 6,6,6,6,6,6,6,5],
    'Total' : [19, 22, 23,15,15,19,13,18,22,27]
}
data = pd.DataFrame(data) 

# bar chart 1 -> top bars (group of 'smoker=No')
bar1 = sns.barplot(x="Player",  y="Total", data=data, color='#7a2048')

# bar chart 2 -> bottom bars (group of 'smoker=Yes')
bar2 = sns.barplot(x="Player", y="Player Goals/Team Goals", data=data, color='#408ec6')
bar2.set(xlabel=None)
def addlabels(x,y):
    for i in range(len(x)):
        bar1.text(i,y[i],y[i],verticalalignment='bottom',fontsize ='medium',ha="center",size=15)
addlabels(data['Player'], data['Total'])
def addlabels_(x,y):
    for i in range(len(x)):
        bar2.text(i,y[i],y[i],verticalalignment='bottom',fontsize ='medium',ha="center",size=15,color='white')
addlabels_(data['Player'], data['Player Goals/Team Goals'])
# add legend
top_bar = mpatches.Patch(color='darkblue', label='smoker = No')
bottom_bar = mpatches.Patch(color='lightblue', label='smoker = Yes')


# show the graph
plt.savefig("Goals.png")
plt.show()
```
