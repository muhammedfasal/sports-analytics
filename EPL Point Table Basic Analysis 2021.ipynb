```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_html('https://www.bbc.co.uk/sport/football/premier-league/table')
df = df[0]
```

```python
df
```

```python
df.info()
```

```python
df.rename(columns={'Unnamed: 0': 'Position'}, inplace=True)
df.drop('Unnamed: 1', inplace=True, axis=1)
df.drop('Form', inplace=True, axis=1)
df.drop(df.tail(1).index, inplace=True)
```

```python
df
```

```python
df[['Position','P','W','D','L','F','A','GD','Pts']] = df[['Position','P','W','D','L','F','A','GD','Pts']].apply(pd.to_numeric,errors='coerce')
```

```python
df.info()
```

```python
plt.bar(df["Position"],height=df['Pts'])
```

```python
plt.bar(x=np.arange(1,21),height=df['Pts'])
plt.title("Premier League 20/21")
plt.xticks(np.arange(1,21), df['Team'], rotation=90)
plt.xlabel("Team")
plt.ylabel("Points")
plt.show()
```

```python
plt.bar(x=df["Position"],height=df['GD'])
plt.title("Premier League 20/21")
plt.xticks(np.arange(1,21), df['Team'], rotation=90)
plt.xlabel("Team")
plt.ylabel("GD")
plt.show()
```

```python
plt.bar(x=df["Pts"],height=df['GD'])
plt.title("Premier League 20/21")
plt.xticks(df["Pts"], df['Team'], rotation=90)
plt.xlabel("Team")
plt.ylabel("GD")
plt.show()
```

```python
plt.hlines(y=np.arange(1,21),xmin=0,xmax=df['Pts'])
```

```python
plt.hlines(y=np.arange(1,21),xmin=0,xmax=df['Pts'],color="skyblue")
plt.plot(df['Pts'], np.arange(1,21), "o")
plt.yticks(np.arange(1,21), df['Team'])
plt.show()
```

```python
plt.plot(df['F'],df['A'],"o")
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(7, 5)
plt.plot(df['F'],df['A'],"*")
ax.set_title("Goals for & Against")
ax.set_xlabel("Goals For")
ax.set_ylabel("Goals Against")
for i, txt in enumerate(df['Team']):
    ax.annotate(txt, (df['F'][i], df['A'][i]))
```
